```python
df_clean["nature_mutation"].value_counts()
```

```python
df_clean= df_clean[df["nature_mutation"] == "Vente"]
df_clean["nature_mutation"].value_counts()
```

```python
df_clean = df_clean[df_clean["type_local"].isin(["Maison", "Appartement"])]

df_clean["type_local"].value_counts()
```

```python
carrez_cols = [
    "lot1_surface_carrez",
    "lot2_surface_carrez",
    "lot3_surface_carrez",
    "lot4_surface_carrez",
    "lot5_surface_carrez"
]

df_clean["surface_carrez_totale"] = df_clean[carrez_cols].sum(axis=1, skipna=True)
```

```python
lot_cols = [
    "lot1_numero",
    "lot2_numero",
    "lot3_numero",
    "lot4_numero",
    "lot5_numero"
]

df_clean["nb_lots_renseignes"] = df_clean[lot_cols].notna().sum(axis=1)
```

```python
df_clean[["surface_carrez_totale", "nb_lots_renseignes"]].head()
```

```python
df_clean = df_clean.drop(columns=carrez_cols + lot_cols)
```
